# What is the difference between a function and a method in Python?

--> Function: A function is an independent block of reusable code that performs a specific task.

It is defined using the def keyword.

It can take arguments and return values.

It can be called anywhere in the program.

Method: A method is a function that is associated with an object (i.e., it belongs to a class).

It is defined inside a class.

It takes self as its first parameter (in instance methods), which refers to the instance of the class.

It is called on an object of the class.

# Explain the concept of function arguments and parameters in Python.

--> In Python, function arguments and parameters allow data to be passed into functions, making them more dynamic and reusable. Parameters are the variables listed in a function definition, while arguments are the actual values passed when calling the function. Python supports different types of arguments, including positional arguments, which must be passed in the correct order, and default arguments, which take a predefined value if no argument is provided. Additionally, keyword arguments allow passing values using parameter names, making the order irrelevant. Python also supports arbitrary arguments with *args, which collects multiple positional arguments into a tuple, and **kwargs, which gathers multiple keyword arguments into a dictionary. These different argument types can be combined in a specific order to create flexible functions. By using them effectively, Python functions can handle a variety of inputs, improving code modularity and reusability.

# What are the different ways to define and call a function in Python?

--> Different ways to define and call a function in Python are:

1. Regular Function: A function that is defined using the def keyword and can take arguments to perform a specific task.

2.  Function with Default Arguments: A function where one or more parameters have default values, which are used if no argument is passed during the function call.

3. Function with *args (Arbitrary Positional Arguments): A function that allows passing a variable number of positional arguments, which are collected into a tuple.

4. Function with **kwargs (Arbitrary Keyword Arguments): A function that allows passing a variable number of named (keyword) arguments, which are collected into a dictionary.

5.  Function with **kwargs (Arbitrary Keyword Arguments): A function that allows passing a variable number of named (keyword) arguments, which are collected into a dictionary.

6. Function as an Argument: A function that takes another function as an argument and applies it to a value.

7.  Function Returning Another Function: A function that returns another function as its result.

8. Recursive Function: A function that calls itself to solve problems like factorial or Fibonacci sequences.

9. Nested Function (Function Inside Another Function): A function that is defined inside another function, used for better encapsulation.

10. Method in a Class: A function that is defined inside a class and operates on class instances.


# What is the purpose of the `return` statement in a Python function?

--> The return statement in Python is used to send back a value from a function to its caller, marking the end of the function's execution. It allows functions to process data and return results, which can then be stored in variables or used in further computations. A function can return a single value, multiple values as a tuple, or even another function. If no return statement is provided, Python implicitly returns None. Additionally, return can be used to exit a function early based on a condition. Unlike print(), which only displays output, return allows functions to produce reusable results, making them more versatile and efficient.

#  What are iterators in Python and how do they differ from iterables?

--> An iterable is any Python object that can be looped over (e.g., lists, tuples, strings, dictionaries, sets). It contains data that can be accessed one element at a time using a loop. Iterables implement the __iter__() method, which returns an iterator.

--> An iterator is an object that produces one element at a time from an iterable when called with next(). It remembers its state (i.e., where it is in the sequence). Iterators implement both __iter__() (to return itself) and __next__() (to get the next value).

Differences Between Iterables and Iterators in Python:-

Methods Implemented:

An iterable must have the __iter__() method, which returns an iterator whereas in iterator must have both __iter__() and __next__() methods.

State Storage:

An iterable does not store the state of iteration; it can be looped over multiple times.

An iterator remembers its position in the sequence and continues from where it left off.

Usage:

Iterables can be directly used in loops like for loops.

Iterators require calling next() explicitly to get the next element.



# Explain the concept of generators in Python and how they are defined.

--> Generators in Python are a special type of iterator that allow lazy evaluation, meaning they generate values on demand instead of storing them in memory all at once. They are useful for handling large datasets efficiently. Unlike normal functions, which return a single value using return, generators use the yield keyword to produce a sequence of values.

How Generators Work:

1. Defined using def and yield instead of return.

2. Produce values one at a time instead of returning all at once.

3. Automatically save their state and resume from the last execution when next() is called.

4. Efficient in terms of memory because they don’t store all values in memory like lists.

# What are the advantages of using generators over regular functions?

--> 1. Memory Efficiency (Lazy Evaluation): Generators do not store all values in memory like lists. Instead, they generate values on demand, making them ideal for working with large datasets. In contrast, regular functions that return lists consume a lot of memory, especially for large sequences.

2. ⚡ Faster Execution (No Extra Storage Overhead): Since generators yield values one at a time, they avoid the overhead of storing large lists in memory. This allows them to start producing results immediately, while regular functions must compute everything before returning.

3.  Infinite Sequence Handling: Generators can generate infinite sequences without running out of memory, something that regular functions cannot do. If a function tries to return an infinite list, it will eventually cause a crash due to excessive memory usage.

4. leaner and More Readable Code: Generators allow simpler and more efficient iteration, making code more readable compared to using manual loops or handling large lists. A common example is reading large files line by line without loading the entire file into memory.

#  Explain the purpose and usage of the `map()` function in Python.

--> The map() function in Python is used to apply a function to each item in an iterable (such as a list or tuple) and return a new iterable (a map object). It allows for efficient element-wise transformations without needing explicit loops.

When to Use map():-
When you need to apply the same function to every element in a list or tuple.
When you want to avoid loops for better performance and readability.
When dealing with large datasets, since map() is memory efficient (returns an iterator).

#  What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

--> Map() - Applies a Function to Each Element
Purpose: Transforms each element in an iterable using a function.

Returns: A map object (an iterator) with transformed values.

Use Case: When you need to modify all elements in a sequence.

filter() – Selects Elements That Meet a Condition
Purpose: Filters elements from an iterable based on a condition.

Returns: A filter object (an iterator) with only elements that satisfy the condition.

Use Case: When you need to keep or remove elements based on a test.

reduce() – Reduces an Iterable to a Single Value
Purpose: Performs a rolling computation on an iterable, combining elements using a function.

Returns: A single accumulated result.

Use Case: When you need to compute a single value from a sequence (e.g., sum, product).

```python
#  Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.
def square_numbers(numbers):
    return list(map(lambda x: x ** 2, numbers))

num_list = [1, 2, 3, 4, 5]
result = square_numbers(num_list)
print(result)
```

```python
# Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

prime_numbers = [n for n in range(1, 201) if is_prime(n)]
print(prime_numbers)

```

```python
# Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class FibonacciIterator:
    def __init__(self, n_terms):
        self.n_terms = n_terms
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n_terms:
            raise StopIteration
        fib_number = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return fib_number

fib_iterator = FibonacciIterator(10)
for num in fib_iterator:
    print(num, end=" ")
```

```python
#  Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_two(max_exponent):
    for exp in range(max_exponent + 1):
        yield 2 ** exp

for num in powers_of_two(5):
    print(num, end=" ")
```

```python
#  Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_line_by_line(file_path):
    """Generator function to read a file line by line."""
    with open(file_path, "r") as file:
        for line in file:
            yield line.strip()  # Yield each line without leading/trailing spaces

file_path = "Happy.text"  # Replace with your file name
for line in read_file_line_by_line(file_path):
    print(line)

```

```python
#  Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

data = [(1, 5), (3, 2), (4, 8), (2, 1)]

sorted_data = sorted(data, key=lambda x: x[1])

print(sorted_data)
```

```python
# Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

celsius_temps = [0, 10, 20, 30, 40, 100]

fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))

print("Cel:", celsius_temps)
print("Fah:", fahrenheit_temps)
```

```python
#  Create a Python program that uses `filter()` to remove all the vowels from a given string.

def remove_vowels(string):
    vowels = "aeiouAEIOU"  # Define vowels
    return "".join(filter(lambda char: char not in vowels, string))

text = "Hello, World!"
result = remove_vowels(text)

print("Original String:", text)
print("String without vowels:", result)
```
